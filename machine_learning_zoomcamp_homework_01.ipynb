```python
asia = df1[df1["origin"] == "Asia"][["vehicle_weight", "model_year"]]
```

```python
asia7 = asia.head(7)
asia7
```

```python
X = asia7.to_numpy()
```

```python
XTX = X.T @ X
print(XTX)
XTX_inv = np.linalg.inv(XTX)
print(XTX_inv)
```

```python
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
```

```python
w = XTX_inv @ X.T @ y
print(w)
```
